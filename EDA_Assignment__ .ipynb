```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from IPython.display import display
from sklearn import datasets
import scipy.stats as stats  
```

```python
df= pd.read_csv('HRDataset_v14.csv')
df.head()
```

```python
df.columns
```

### Univariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
RecruitmentSource_Order = df['RecruitmentSource'].value_counts().index
sns.countplot(data=df, y= "RecruitmentSource",order=RecruitmentSource_Order,
              color = 'g')
plt.xticks(rotation=90)
plt.show()
```

#### Most of Recruitment Source was from indeed.
#### Technology and social media have become a powerful cause of employment.

```python
Department_Order = df['Department'].value_counts().index
sns.countplot(data=df, y= "Department",order=Department_Order,
              color = 'g')
plt.xticks(rotation=90)
plt.show()
```

##### Most of employees is Production Department.
##### Employees whose Production Department are 15 times more than Admin Executive Office Department.


### Bivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
ax = sns.countplot(x="Department", hue="Termd", data=df )

loc, labels = plt.xticks()
ax.set_xticklabels(labels, rotation=90);
plt.show()
```

#### There are many employees who left the company in the production department.

```python

sns.countplot(data= df, y='RecruitmentSource', hue ='Sex')
plt.xticks(rotation=30)
plt.show()
```

#### -In Employee Referral, the number of men is twice the number of women.. I think men have strong communication skills.
